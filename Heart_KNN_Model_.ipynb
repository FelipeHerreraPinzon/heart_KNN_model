#**Heart Dataset K Neighbors Classifier ðŸ’—ðŸ«€**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Cargar datos desde los archivos CSV
X_train = pd.read_csv('X_train_heart.csv')
y_train = pd.read_csv('y_train_heart.csv')

# Eliminar filas con NaN de ambos conjuntos de datos
X_train_clean = X_train.dropna(axis=0)
y_train_clean = y_train.dropna(axis=0)

# Asegurar que las filas eliminadas de X_train coincidan con las filas eliminadas de y_train
y_train_clean = y_train_clean.iloc[X_train_clean.index]

# Convertir y_train_clean a un array unidimensional
y_train_clean = y_train_clean.values.ravel()

# Dividir el conjunto de datos limpio en entrenamiento y prueba
X_train_final, X_test, y_train_final, y_test = train_test_split(X_train_clean, y_train_clean, test_size=0.3, random_state=42)

# Crear un clasificador k-NN
knn_heart_model = KNeighborsClassifier(n_neighbors=1)

# Entrenar el clasificador con el conjunto de entrenamiento
knn_heart_model.fit(X_train_final, y_train_final)

# Predecir las etiquetas del conjunto de prueba
y_pred = knn_heart_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("PrecisiÃ³n del modelo k-NN:", accuracy)
```

#**Heart Dataset KNNImputer ðŸ’—ðŸ«€**

```python
from sklearn.impute import KNNImputer
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Cargar datos desde los archivos CSV
X_train = pd.read_csv('X_train_heart.csv')
y_train = pd.read_csv('y_train_heart.csv')

# Crear un objeto imputer KNN
imputer = KNNImputer(n_neighbors=100)

# Imputar los valores faltantes en X_train (conjunto de datos con caracterÃ­sticas)
X_train_imputed = imputer.fit_transform(X_train)

# Convertir y_train a un array unidimensional
y_train = y_train.values.ravel()

# Dividir el conjunto de datos imputado en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_train_imputed, y_train, test_size=0.3, random_state=42)

# Crear un clasificador k-NN
knn_heart_model = KNeighborsClassifier(n_neighbors=80)

# Entrenar el clasificador con el conjunto de entrenamiento imputado
knn_heart_model.fit(X_train, y_train)

# Predecir las etiquetas del conjunto de prueba
y_pred = knn_heart_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("PrecisiÃ³n del modelo k-NN:", accuracy)

```

#**Heart Dataset Random Forest Classifier ðŸ’—ðŸ«€**

```python
from sklearn.impute import KNNImputer
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Cargar datos desde los archivos CSV
X_train = pd.read_csv('X_train_heart.csv')
y_train = pd.read_csv('y_train_heart.csv')

# Crear un objeto imputer KNN
imputer = KNNImputer(n_neighbors=30)

# Imputar los valores faltantes en X_train (conjunto de datos con caracterÃ­sticas)
X_train_imputed = imputer.fit_transform(X_train)

# Convertir y_train a un array unidimensional
y_train = y_train.values.ravel()

# Dividir el conjunto de datos imputado en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_train_imputed, y_train, test_size=0.3, random_state=42)

# Crear un clasificador Random Forest
random_forest_model = RandomForestClassifier()

# Entrenar el clasificador con el conjunto de entrenamiento imputado
random_forest_model.fit(X_train, y_train)

# Predecir las etiquetas del conjunto de prueba
y_pred = random_forest_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("PrecisiÃ³n del modelo Random Forest:", accuracy)
```

#**Heart Dataset Decision Tree Classifier ðŸ’—ðŸ«€**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Cargar datos desde los archivos CSV
X_train = pd.read_csv('X_train_heart.csv')
y_train = pd.read_csv('y_train_heart.csv')

# Eliminar filas con NaN de ambos conjuntos de datos
X_train_clean = X_train.dropna(axis=0)
y_train_clean = y_train.dropna(axis=0)

# Asegurar que las filas eliminadas de X_train coincidan con las filas eliminadas de y_train
y_train_clean = y_train_clean.iloc[X_train_clean.index]

# Convertir y_train_clean a un array unidimensional
y_train_clean = y_train_clean.values.ravel()

# Dividir el conjunto de datos limpio en entrenamiento y prueba
X_train_final, X_test, y_train_final, y_test = train_test_split(X_train_clean, y_train_clean, test_size=0.3, random_state=42)

# Crear un clasificador de Ã¡rbol de decisiÃ³n
decision_tree_model = DecisionTreeClassifier()

# Entrenar el clasificador con el conjunto de entrenamiento
decision_tree_model.fit(X_train_final, y_train_final)

# Predecir las etiquetas del conjunto de prueba
y_pred = decision_tree_model.predict(X_test)

# Calcular la precisiÃ³n del modelo
accuracy = accuracy_score(y_test, y_pred)
print("PrecisiÃ³n del modelo de Ã¡rbol de decisiÃ³n:", accuracy)
```
